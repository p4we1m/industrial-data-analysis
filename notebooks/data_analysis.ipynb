```python
import pandas as pd
import matplotlib.pyplot as plt

# Wczytaj dane
df = pd.read_csv("synthetic_machine_data.csv", parse_dates=["timestamp"])

# Analiza trendw
plt.figure(figsize=(10, 6))
df["temperature"].plot(label="Temperature", alpha=0.7)
df["vibration"].plot(label="Vibration", alpha=0.7)
plt.legend()
plt.title("Trendy danych")
plt.show()

# Wykrywanie anomalii
anomalies = df[df["temperature"] > 80]
print(f"Liczba anomalii temperatury: {len(anomalies)}")
```
